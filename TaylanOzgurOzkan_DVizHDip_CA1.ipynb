### 2.2. Correlation Analysis: NA and JP Sales (2010-2014)

In this step, we will examine the correlation between North America (NA) and Japan (JP) sales for games released between 2010 and 2014. Correlation indicates the strength and direction of the relationship between two variables. The goal here is to understand whether there are similar sales trends between the two regions. After analyzing the correlation, we will create a scatter plot as a visualization and indicate the correlation coefficient.

```python
# Correlation of North America (NA) and Japan (JP) sales from 2010-2014
filtered_df = df[(df['Year_of_Release'] >= 2010) & (df['Year_of_Release'] <= 2014)]
na_jp_corr = filtered_df['NA_Sales'].corr(filtered_df['JP_Sales'])
na_jp_corr
```

A positive but weak correlation (0.26) was found between North American and Japanese sales between 2010 and 2014. This correlation suggests that sales in the two markets do not strongly influence each other. This may indicate that the two markets may have different gaming preferences and demographics. Companies should develop separate strategies for the two markets.

```python
# Visualization with scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df["NA_Sales"], filtered_df["JP_Sales"], alpha=0.5, color="blue")
plt.title("Correlation between NA and JP Sales (2010-2014)", fontsize=14)
plt.xlabel("NA Sales (Millions)", fontsize=12)
plt.ylabel("JP Sales (Millions)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The correlation between North America (NA) and Japan (JP) sales between 2010-2014 was calculated as 0.26. This correlation value indicates a weak positive relationship. The scatter plot visually presents the relationship between sales. The scattered dots indicate that there is no strong dependency between these two regions; that is, just because a game is popular in North America does not guarantee that it will be popular in Japan.

```python
# Histogram of NA sales
plt.figure(figsize=(10, 6))
plt.hist(filtered_df["NA_Sales"], bins=20, color="green", alpha=0.7, edgecolor="black")
plt.title("NA Sales Distribution (2010-2014)", fontsize=14)
plt.xlabel("NA Sales (Millions)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Histogram of JP sales
plt.figure(figsize=(10, 6))
plt.hist(filtered_df["JP_Sales"], bins=20, color="orange", alpha=0.7, edgecolor="black")
plt.title("JP Sales Distribution (2010-2014)", fontsize=14)
plt.xlabel("JP Sales (Millions)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The histograms of NA and JP sales show the sales concentrations between the two markets. The vast majority of sales in North America are concentrated in the 0-5 million range, while sales in Japan are generally concentrated at lower levels, rarely exceeding 1 million sales. This suggests that the North American market has a larger volume of game sales than Japan.

```python
import seaborn as sns

# Heatmap for distribution of NA and JP sales
plt.figure(figsize=(10, 6))
sns.histplot(x=filtered_df["NA_Sales"], y=filtered_df["JP_Sales"], bins=30, pmax=0.8, cmap="Greens")
plt.title("Heatmap of NA vs JP Sales (2010-2014)", fontsize=14)
plt.xlabel("NA Sales (Millions)", fontsize=12)
plt.ylabel("JP Sales (Millions)", fontsize=12)
plt.tight_layout()
plt.show()
```

### 2.4. Comparison of European Sales for Old Games (2005 and Earlier) and New Games (After 2005)

In this step, we will compare the average European sales of games released before 2005 with those released after 2005. This analysis will help us understand how old and new games perform in Europe. We will calculate the average European sales by splitting the data into two groups and then visualize this comparison. We will use three different visualizations (bar chart, box plot and violin plot) to look at it from different perspectives.

```python
# Splitting the dataset into two for games from 2005 and before and for games from 2005 onwards
older_games = df[df["Year_of_Release"] <= 2005]
newer_games = df[df["Year_of_Release"] > 2005]

# Calculating the average of European sales of both groups
mean_eu_sales_older = older_games["EU_Sales"].mean()
mean_eu_sales_newer = newer_games["EU_Sales"].mean()

# Print average sales
mean_eu_sales_older, mean_eu_sales_newer
```

The average European sales of games released before 2005 were calculated as 0.148 million, and the average European sales of games released after 2005 were calculated as 0.145 million. There is a very small difference in European sales between these two groups and this difference may not be statistically significant.

First, we will visualize the difference between two groups with a bar chart. Bar charts are very effective for comparing the average sales of two groups.

```python
# Comparing average European sales with a bar chart
plt.figure(figsize=(8, 6))
plt.bar(["2005 and Earlier", "After 2005"], [mean_eu_sales_older, mean_eu_sales_newer], color=["blue", "green"])
plt.title("Average EU Sales: 2005 and Earlier vs. After 2005", fontsize=14)
plt.xlabel("Game Release Year", fontsize=12)
plt.ylabel("Average EU Sales (Millions)", fontsize=12)
plt.tight_layout()
plt.show()
```

The bar chart shows that the average European sales of games released before 2005 were slightly higher than those released after 2005. However, this difference is quite small. Visually, it can be observed that there is no significant difference between the two groups.

Now, let's examine the distribution of European sales using a boxplot. This plot will show how sales are distributed for both groups, as well as any possible outliers.

```python
# Comparison of old and new games' European sales with box charts
plt.figure(figsize=(10, 6))
sns.boxplot(data=[older_games["EU_Sales"], newer_games["EU_Sales"]], palette=["blue", "green"])
plt.xticks([0, 1], ["2005 and Earlier", "After 2005"])
plt.title("Boxplot of EU Sales: 2005 and Earlier vs. After 2005", fontsize=14)
plt.xlabel("Game Release Year", fontsize=12)
plt.ylabel("EU Sales (Millions)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The boxplot shows that European sales of games released in 2005 and earlier are distributed over a wide range. It can also be observed that while the average sales for both groups are similar, some games have achieved particularly high sales figures. This plot provides a clearer picture of the distribution of European sales for both groups and possible outliers.

Finally, we will examine the sales distribution of both groups with the violin plot. This graph provides more information than the box plot, showing the sales density and distribution of both groups more clearly.

```python
# Comparing the distribution of European sales with the Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=[older_games["EU_Sales"], newer_games["EU_Sales"]], palette=["blue", "green"])
plt.xticks([0, 1], ["2005 and Earlier", "After 2005"])
plt.title("Violin Plot of EU Sales: 2005 and Earlier vs. After 2005", fontsize=14)
plt.xlabel("Game Release Year", fontsize=12)
plt.ylabel("EU Sales (Millions)", fontsize=12)
plt.tight_layout()
plt.show()
```

The violin plot shows the density and distribution of sales in both periods in more detail. The graph shows that sales of games from 2005 and before are distributed in a wider range, but average sales are at similar levels for both periods. In addition, thanks to the violin plot, it is possible to clearly observe the regions where sales are concentrated at certain values.

With these three different visualizations, we examined European sales of games released before 2005 and games released after 2005 from different perspectives:

The bar chart clearly showed the average sales difference between the two groups.
The box plot showed the distribution of sales and possible outliers.
The violin plot allowed us to see the density and distribution of sales in more detail.

Step 2: Relationship Between User Ratings and Global Sales

Now, we will analyze the impact of user ratings on global sales. We will calculate the correlation and visualize it with a scatter plot to see if games with higher user ratings sell more.

```python
# Let's check the data types of the User_Score and Global_Sales columns
print(df["User_Score"].dtype)
print(df["Global_Sales"].dtype)
```

```python
# Let's do the same for Global_Sales
df["Global_Sales"] = pd.to_numeric(df["Global_Sales"], errors="coerce")
```

```python
# Let's check the data types again
print(df["User_Score"].dtype)
print(df["Global_Sales"].dtype)
```

```python
# Converting user ratings to a numeric type
df["User_Score"] = pd.to_numeric(df["User_Score"], errors="coerce")

# Recalculating the correlation between user score and global sales
user_correlation = df["User_Score"].corr(df["Global_Sales"])

# Print correlation result
user_correlation
```

Final Visualization: Relationship Between Reviewer Rating, User Rating, and Reviewer Count and Global Sales

In this step, we will combine the relationships between reviewer rating, user rating, and reviewer count and global sales into a single visual. We will visualize which variables correlate with global sales by comparing these three variables with different colors on scatter plots.

```python
# Visualize the relationship between reviewer rating, user rating, and number of reviewers and global sales in one chart
plt.figure(figsize=(10, 6))

# Critic Score and global sales
plt.scatter(df["Critic_Score"], df["Global_Sales"], alpha=0.5, color="blue", label="Critic Score")

# User Score and global sales
plt.scatter(df["User_Score"], df["Global_Sales"], alpha=0.5, color="green", label="User Score")

# Critic Count and global sales
plt.scatter(df["Critic_Count"], df["Global_Sales"], alpha=0.5, color="red", label="Critic Count")

plt.title("Critic Score, User Score and Critic Count vs Global Sales", fontsize=14)
plt.xlabel("Score / Count", fontsize=12)
plt.ylabel("Global Sales (Millions)", fontsize=12)
plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

This graph presents the relationships between critic ratings (blue), user ratings (green), and number of critics (red) and global sales in a single visual. While critic counts have a stronger correlation with global sales than critic ratings, user ratings have a much weaker effect on sales. Games with a higher number of critics tend to sell more.
